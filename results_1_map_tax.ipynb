# Automatic mapping

```python
labels_arr[0].shape, predictions[0].shape
```

```python
complete_labels = np.concatenate(labels_arr)
complete_predictions = np.concatenate(predictions)

complete_labels.shape, complete_predictions.shape
```

```python
n_labels = complete_labels.shape[1]
n_pred_labels = complete_predictions.shape[1]

similarity_matrix = np.zeros((n_pred_labels, n_labels))
for i in range(n_pred_labels):
    for j in range(n_labels):
        # the similarity is the dot product divided by the product of the norms
        if np.linalg.norm(complete_predictions[:, i]) * np.linalg.norm(complete_labels[:, j]) == 0:
            similarity_matrix[i, j] = 0
        else:
            similarity_matrix[i, j] = np.dot(complete_predictions[:, i], complete_labels[:, j]) / (np.linalg.norm(complete_predictions[:, i]) * np.linalg.norm(complete_labels[:, j]))

similarity_matrix.shape
```

```python
import plotly.express as px

fig = px.imshow(similarity_matrix, aspect='auto', x=taxonomy_names, y=yamnet_names, color_continuous_scale='Inferno')
fig.show()
```

```python
automatic_map = {}

for i, tax_name in enumerate(yamnet_names):
    best_match = np.argmax(similarity_matrix[i, :])
    automatic_map[i] = best_match

    similarity_val = similarity_matrix[i, best_match]
    best_match_name = taxonomy_names[best_match]
    if similarity_val != 0:
        print(f"{tax_name} -> {best_match_name}. Similarity: {similarity_val}")
```

```python
auto_mapped_predictions = map_predictions(predictions, automatic_map)

true_y = np.concatenate(labels_arr)
auto_pred_y = np.concatenate(auto_mapped_predictions)

evaluate(true_y, auto_pred_y, threshold=0.5)
```

```python
for threshold in [0.1, 0.3, 0.5, 0.7, 0.9]:
    print(f"Threshold: {threshold}")
    evaluate(true_y, auto_pred_y, threshold=threshold, plot_confusion_matrices=False)
```
