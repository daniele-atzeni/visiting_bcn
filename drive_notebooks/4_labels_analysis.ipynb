```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df_20 = pd.read_csv('data/yamnet_foreground_labels_20.csv')
df_21 = pd.read_csv('data/yamnet_foreground_labels_21.csv')
```

```python
df_20
```

```python
df_21
```

Let's try with the co-occurrencies of labels

```python
unique_lab_20 = np.unique(df_20[[el for el in df_20.columns if 'label_' in el]].values)
unique_pred_20 = np.unique(df_20['predicted_label'].values)

co_occ_20 = np.zeros((len(unique_lab_20), len(unique_pred_20)))

for i, row in df_20.iterrows():
    pred = row.predicted_label
    labs = []
    for col, el in row.items():
        if 'label_' in col and el != 'None':
            labs.append(el)

    for lab in labs:
      co_occ_20[np.where(lab==unique_lab_20), np.where(pred==unique_pred_20)] += 1
```

```python
co_occ_20
```

```python
fig = px.imshow(co_occ_20, aspect='auto', x=unique_pred_20, y=unique_lab_20, color_continuous_scale='Inferno')
fig.show()
```

Let's ceil max values, just to see which labels occur together at least X times

```python
top_val = 250
fig = px.imshow(np.minimum(co_occ_20, top_val), aspect='auto', x=unique_pred_20, y=unique_lab_20, color_continuous_scale='Inferno')
fig.show()
```
